```python
# Import Data
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',header=None)
# Give appropriate header to each columns
data.columns = ['Age','Workclass','fnlwgt','Education','Education-num','Marital-status','Occupation','Relationship','Race','Sex','Capital-gain','Capital-loss','Hours-per-week','Native-country','Income']
# Show the "data"
data.head(10)
```

# Investigate the data

```python
# Print out the Mean, Standard Deviation, Minimum Value, Maximum Value, and Variance for each numerical Feature
from pandas.api.types import is_numeric_dtype

for col in data.columns:
    if is_numeric_dtype(data[col]):
        print('%s:' % (col))
        print('\t Mean = %.2f' % data[col].mean())
        print('\t Standard deviation = %.2f' % data[col].std())
        print('\t Minimum = %.2f' % data[col].min())
        print('\t Maximum = %.2f' % data[col].max())
        print('\t Variance = %.2f' % data[col].var())
```

```python
# Check a more detailed statistics of the data including the numerical and categorical Feature
data.describe(include = 'all')
```

# Visualize the Data

```python
# Plot the Income with the respect of number of occurence
%matplotlib inline

data['Income'].hist(bins=10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure (figsize = (6,6))
sns.boxplot(x = "Income", y ="Age", data = data)
plt.title('Income distribution by Age')
plt.xlabel('Income')
plt.ylabel('Age')
plt.show()
```

<font color= "Red"> Findings:</font> The 40-50's have more income than the 30-40's and it is reasonable as age can affect one's income. the 30-40's might have lower income due to their entry level position compared to the 40-50's which are more stable. 

```python
plt.figure (figsize = (10,6))
sns.countplot(x="Workclass", hue="Income", data=data)
plt.title(f'Income distribution by Workclass')
plt.xlabel('Income')
plt.ylabel('Count')
```

<font color= "Red"> Findings:</font> Thiw show that majority of the workclass is "Private employees" and Private employee has more income than other workclass

```python
plt.figure (figsize = (6,6))
sns.boxplot(x = "Income", y ="fnlwgt", data = data)
plt.title('Income distribution by Final weight')
plt.xlabel('Income')
plt.ylabel('fnlwgt')
plt.show()
```

<font color = 'red'>Findings</font> :The graph has little differences which indicates that fnlgt variable did not really affect income

```python
plt.figure(figsize=(15, 6))
sns.countplot(x="Education", hue="Income", data=data)
plt.title(f'Income distribution by Education')
plt.xlabel("Education")
plt.ylabel('Count')
plt.xticks(rotation=30)
plt.legend(title='Income', labels=['<=50K', '>50K'])
plt.show()
```

<font color = 'red'>Findings :</font> From the graph we can see that the hight education background one have (Bachelors, Masters), the higher income he will get 

```python
plt.figure (figsize = (10,6))
sns.boxplot(x = "Income", y ="Education-num", data = data)
plt.title('Income distribution by Education level')
plt.xlabel('Income')
plt.ylabel('Education level')
plt.show()
```

<font color = 'red'>Findings :</font> Boxplot clearly show the higher education level one has, the higher income he will get

```python
plt.figure(figsize=(15, 6))
sns.countplot(x="Marital-status", hue="Income", data=data)
plt.title(f'Income distribution by Maritas Status')
plt.xlabel("Marital - Status")
plt.ylabel('Count')
plt.xticks(rotation=30)
plt.legend(title='Income', labels=['<=50K', '>50K'])
plt.show()
```
